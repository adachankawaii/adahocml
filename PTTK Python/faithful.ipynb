```python
from sklearn.cluster import KMeans
center = 4
nstart = 25
kmeans = KMeans(n_clusters=center, n_init=nstart)
ff['index'] = kmeans.fit_predict(ff[['eruptions']])
unique_clusters = np.unique(ff['index'])
selected_clusters = np.random.choice(unique_clusters, size=4, replace=False)
cluster_sample = ff[ff['index'].isin(selected_clusters)]
cluster_counts = pd.crosstab(index=cluster_sample['index'], columns='count')
ff[ff['index'] == 1] 
```
